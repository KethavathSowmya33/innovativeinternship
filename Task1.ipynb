```python
#Pass/Fail Analyzer
marks=[45,78,90,33,60]
passcount=len([m for m in marks if m>=50])
failcount=len([m for m in marks if m<50])
print("Totalstudents",len(marks))
print("Passstudents",passcount)
print("Failstudents",failcount)
```

```python
#Simple Data Cleaner
names=["Alice","bob","CHARLIE"]
cleanednames=[name.strip().lower() for name in names]
print("Cleaned Names:",cleanednames)
```

```python
#Message Length Analyzer
messages=["Hi","Welcome to the platform","OK"]
for msg in messages:
  length=len(msg)
  print(f"Message:'{msg}'|length:{length}")
  if length>10:
    print("This message is longer than 10 characters\n")
  else:
    print("This is in the correct length\n")
```
