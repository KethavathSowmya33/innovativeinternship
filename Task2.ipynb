```python
# Unique words in a sentence
sentence=input("Enter sentence: ")
words=sentence.split()
uniquewords=set(words)
print("Unique words count:", len(uniquewords))
print("Unique words:",uniquewords)
```

```python
#Highest salary from employee data
employees={
    "Ravi":75000,
    "Anita":68000,
    "Kiran":72000
}
highest=max(employees,key=employees.get)
print("Employee with highest salary:", highest)
print("Highest salary:",employees[highest])
```

```python
#Find maximum and minimum values
numbers=[45,22,89,10,66]
max=numbers[0]
min=numbers[0]
for num in numbers:
  if num>max:
    max=num
  if num<min:
    min=num
print("List:",numbers)
print("Max:",max)
print("Min:",min)
```

```python
#Count products above a price threshold
prices=[450,1200,899,1500,300]
count=0
for price in prices:
  if price > 1000:
    count+=1
print("Products above 1000:",count)
```

```python
#Calculate attendance percentage
attendance=["P","P","A","P","P"]
presentcount=attendance.count("P")
totaldays=len(attendance)
percentage=(presentcount/totaldays)*100
print("Attendance Percentage",percentage)
```

```python
#Remove duplicate phone numbers
phonenumbers=[9876543210,9123456789,9876543210]
uniquenumbers=set(phonenumbers)
print("Uniue phone numbers:",uniquenumbers)
```

```python
#count character frequency
text="python"
charfrequency={}
for char in text:
  if char in charfrequency:
    charfrequency[char]+=1
  else:
    charfrequency[char]=1
print("Character frequency:",charfrequency)
```
