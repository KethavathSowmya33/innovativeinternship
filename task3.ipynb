```python
#employee performance bonus elgibility
employees={
    "Ravi":92,
    "Anita":88,
    "Kiran":92,
    "surech":85
}
highest=max(employees.values())
top_performers=[]
for name, score in employees.items():
  if score==highest:
    top_performers.append(name)
print(f"Top performers Eligible for Bonus:{','.join(top_performers)}(Score:{highest})")
```

```python
#search query keyword analysis
def keyanalysis(query):
  query=query.lower()
  for ch in ['.',',','!','?']:
    query=query.replace(ch,'')
  words=query.split()
  freq={}
  for word in words:
    freq[word]=freq.get(word,0)+1
  result={word:count for word, count in freq.items() if count>1}
  return result
query="Buy mobile phone buy phone online"
print(keyanalysis(query))
```

```python
#Sensor data validation
def validatesensorreadings(sensorreadings):
  validreadings=[]
  for hour,value in enumerate(sensorreadings):
    if value % 2 == 0:
      validreadings.append((hour,value))
  return validreadings
sensorreadings=[3,4,5,7,8,10,12,5]
print("Valid Sensor readings(Hour, Value):")
print(validatesensorreadings(sensorreadings))
```

```python
#email domain usage analysis
def emaildomainusage(emails):
  domaincount={}
  for email in emails:
    domain=email.split('@')[1]
    domaincount[domain]=domaincount.get(domain,0)+1
  total=len(emails)
  for domain,count in domaincount.items():
    percentage=(count/total)*100
    print(f"{domain}:{int(percentage)}%")
emails=[
    "ravi@gmail.com",
    "anita@yahoo.com",
    "kiran@gmail.com",
    "meena@yahoo.com"
]
emaildomainusage(emails)
```

```python
#sales spike detection
def detectsalesspikes(sales):
  avgsales=sum(sales)/len(sales)
  spikes=[]
  for day, value in enumerate(sales,start=1):
    if value>avgsales*1.3:
      spikes.append((day,value))
  for day,value in spikes:
    print(f"Day {day}:{value}")
sales=[1200,1500,900,2200,1400,3000]
detectsalesspikes(sales)
```

```python
#duplicate user ID detection
def detectduplicates(user_ids):
  duplicates={}
  for uid in user_ids:
    duplicates[uid]=duplicates.get(uid,0)+1
  for uid,count in duplicates.items():
    if count>1:
      print(f"{uid}-> {count}times")
user_ids=["user1","user2","user1","user3","user1","user3"]
detectduplicates(user_ids)
```
