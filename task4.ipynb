```python
#Mobile Recharge Validation System
def validate_recharge():
  valid_plans={199,299,399,599}
  while True:
    try:
      amount=int(input("Enter recharge amount:"))
      if amount>= 50 and amount in valid_plans:
        print(f"Recharge successful for plan {amount}")
        break
      else:
        print("Invalid recharge. Must be >= 50rs and one of",valid_plans)
    except ValueError:
      print("Please enter a valid number")
validate_recharge()
```

```python
#Inventory Reorder Alert System
def check_inventory(stock_dict):
  for product, quantity in stock_dict.items():
    if quantity<15:
      print(f"{product}:Reorder Alert(Stock={quantity})")
    else:
      print(f"{product}:Stock OK (Stock={quantity})")
inventory={
    "Apples":10,
    "Bananas":25,
    "Oranges":14,
    "Mangoes":30
}
check_inventory(inventory)
```

```python
#Student Result processing system
def process_result(marks):
  total=0
  for mark in marks:
    total += mark
  average=total/len(marks)
  if average>=50:
    print(f"Average={average:.2f}-> Pass")
  else:
    print(f"Average={average:.2f}->Fail")
student_marks=[60,45,70,55]
process_result(student_marks)
```

```python
#cab Fair estimator with retry option
def calculate_fare(distance,peak_hour=False):
  base_fare=50
  per_km=12
  fare=base_fare + (distance*per_km)
  if peak_hour:
    fare *= 1.25
  return fare
def cab_fare_estimator():
  while True:
    try:
      distance=float(input("Enter distance travelled (in km):"))
      peak = input("Is it peak hour? (yes/no):").strip().lower()
      peak_hour=True if peak == "yes" else False
      total_fare=calculate_fare(distance, peak_hour)
      print(f"Total Fare= {total_fare:.2f}")
      break
    except ValuError:
      print("Invalid input. Please enter a valid number for distance")
cab_fare_estimator()
```

```python
# Employee Attendance Eligibility Checker
def check_attendance(attendance_list):
  total_days=len(attendance_list)
  present_days=0
  for day in attendance_list:
    if day=="p":
      present_days +=1
  attendance_percentage = (present_days/total_days)*100
  if attendance_percentage>=75:
    return f" Eligible(Attendance={attendance_percentage:.2f}%)"
  else:
    return f"Not Eligible (Attendance={attendance_percentage:.2f}%)"
attendance=["P", "A", "P", "P", "A", "p", "P", "P", "A", "P"]
print(check_attendance(attendance))
```

```python
# Password Strength checker
def check_password_strength(password):
  if len(password)<8
    return "Weak: Too short"
  if not any(ch.isdigit() for ch in password):
    return "Weak: No digit"
  if not any(ch in "@#$" for ch in password):
    return "Weak: No special cahracter (@,#,$)"
  return "Strong Password"
print(check_password_strength("abc123@"))
print(check_password_strength("securePass99#"))
```
